```python
import pandas as pd
file_path = "/content/drive/MyDrive/Dataset/Titanic-Dataset.csv"
df = pd.read_csv(file_path)

# Display basic info
print("Titanic Dataset Info:\n")
print(df.info())
print("\nFirst 5 rows of the dataset:\n")
print(df.head())

# Part 1: Percent of total fare revenue by class

# 1. Total Fare across all passengers
total_fare = df['Fare'].sum()

# 2. Total Fare by class
fare_first_class = df[df['Pclass'] == 1]['Fare'].sum()
fare_second_class = df[df['Pclass'] == 2]['Fare'].sum()
fare_third_class = df[df['Pclass'] == 3]['Fare'].sum()

# 3. Combine totals into a list
fare_totals = [fare_first_class, fare_second_class, fare_third_class]

# 4. Calculate proportion of fare revenue by class
fare_percent = [round(f / total_fare * 100, 2) for f in fare_totals]

# 5. Display result
print("\n--- Percentage of Total Fare Revenue by Class ---")
print(f"First Class: {fare_percent[0]}%")
print(f"Second Class: {fare_percent[1]}%")
print(f"Third Class: {fare_percent[2]}%")

# Part 2: Percent of passengers by age group

# 1. Create age group column
def age_group(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

# Handle missing ages by keeping them as NaN
df['Age_group'] = df['Age'].apply(lambda x: age_group(x) if pd.notnull(x) else 'unknown')

# 2. Total number of passengers
total_passengers = df.shape[0]

# 3. Count of passengers in each age group (excluding 'unknown')
age_group_counts = df[df['Age_group'] != 'unknown']['Age_group'].value_counts()

# 4. Calculate proportion
age_group_percent = (age_group_counts / total_passengers * 100).round(2)

# 5. Display result
print("\n--- Percentage of Passengers by Age Group ---")
for group, percent in age_group_percent.items():
    print(f"{group.capitalize()}: {percent}%")
```
